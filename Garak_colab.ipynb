```python
# @title 3. Test our setup
# We can run a super simple self-contained test to check that garak's core code is running OK.
!python -m garak --model_type test.Blank --probes test.Test
```

We can see that garak ran OK. It loaded a generator called Blank, which is a test generator that always returns a blank string, "". test.Blank is the automatically-used default generator whenever the test model is used. Then, garak queued up and ran just one probe, test.Blank, which sends blank strings to the generator. So, test.Blank sent empty strings to a generator that always returns empty strings. These outputs were evaluated using the always.Pass detector, which (as you can guess from its name) returned a Pass regardless of the output it was assessing. The final score was 10/10, a pass.



```python
# @title 4. Set Up Your LLM (Generator)
# Before scanning, you need to tell Garak which LLM you want to test.
# This example uses a Hugging Face model that can run locally on Colab's GPU (if available).
# You can also use API-based models like OpenAI's GPT, but you'll need an API key.

# --- Choose your LLM ---

# Option 1: Hugging Face local model (recommended for easy Colab setup)
# This will download the model to your Colab instance.
MODEL_TYPE = "huggingface"
MODEL_NAME = "openai-community/gpt2" # A small, easily loadable model for demonstration
# You can try other models, e.g., "stabilityai/stablelm-tuned-alpha-3b"
# or larger models if you have Colab Pro.

# Option 2: OpenAI GPT (requires API key)
# Uncomment and replace "YOUR_OPENAI_API_KEY" with your actual key.
# MODEL_TYPE = "openai"
# MODEL_NAME = "gpt-3.5-turbo"
# import os
# os.environ["OPENAI_API_KEY"] = "YOUR_OPENAI_API_KEY"
# from google.colab import userdata
# try:
#     os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
# except Exception as e:
#     print(f"Could not load OpenAI API key from Colab secrets: {e}")
#     print("Please set the OPENAI_API_KEY environment variable or Colab secret.")


print(f"Configured to use model: {MODEL_NAME} (Type: {MODEL_TYPE})")
```

```python
# @title 4. Run a Garak Scan
# Now we'll run a scan using the configured LLM and a selected probe.
# For simplicity, we'll use a common jailbreak probe like 'dan' (Do Anything Now).
# You can choose other probes from the list generated in Step 2.

# PROBE = "dan" # A common jailbreak probe
# PROBE = "promptinject" # Another common vulnerability
PROBE = "lmrc.Profanity" # To check for profanity generation
# PROBE = "all" # To run all probes (can be very time-consuming!)

# Define a report prefix for your scan results
REPORT_PREFIX = f"garak_scan_{MODEL_NAME.replace('/', '_')}_{PROBE}"

print(f"Starting Garak scan on {MODEL_NAME} with probe: {PROBE}...")

!python -m garak \
    --model_type {MODEL_TYPE} \
    --model_name {MODEL_NAME} \
    --probes {PROBE} \
    --report_prefix {REPORT_PREFIX} \
    --verbose # Optional: for more detailed output during the scan

print("\nScan completed! Results are being processed.")
```

```python
# @title 5. View Scan Results
# Garak generates results in JSONL format.
# We'll read the report.jsonl file and display a summary.

import json
import pandas as pd
import glob

FILE_PREFIX = f"/root/.local/share/garak/garak_runs/{REPORT_PREFIX}"

# Find the latest report file based on the prefix
report_files = glob.glob(f"{FILE_PREFIX}.report.jsonl")
if not report_files:
    print(f"No report file found with prefix: {REPORT_PREFIX}")
else:
    # Sort to get the most recent one if multiple exist
    report_file = sorted(report_files)[-1]
    print(f"Loading report from: {report_file}")

    results = []
    with open(report_file, 'r') as f:
        for line in f:
            entry = json.loads(line)
            if entry.get("entry_type") == "eval":
                results.append(entry)

    if results:
        df = pd.DataFrame(results)
        # Extract relevant info from 'result' dictionary if it exists
        if 'result' in df.columns:
            df_results = pd.json_normalize(df['result'])
            df = pd.concat([df.drop(columns=['result']), df_results], axis=1)

        print("\n--- Garak Scan Summary ---")
        display(df[['probe', 'passed', 'detector']])
        # You might want to display other columns based on the probe's output

        print("\n--- Detailed Hits (if any) ---")
        # Find the hitlog file
        hitlog_files = glob.glob(f"{FILE_PREFIX}.hitlog.jsonl")
        if hitlog_files:
            hitlog_file = sorted(hitlog_files)[-1]
            hits = []
            with open(hitlog_file, 'r') as f_hit:
                for line_hit in f_hit:
                    hits.append(json.loads(line_hit))
            if hits:
                df_hits = pd.DataFrame(hits)
                display(df_hits)
            else:
                print("No specific vulnerabilities ('hits') recorded in the hitlog.")
        else:
            print("No hitlog file found.")

    else:
        print("No evaluation results found in the report. This might indicate an issue with the scan or model.")

# Optional: Download the full report files if you want to inspect them locally
# from google.colab import files
# files.download(report_file)
# if hitlog_files:
#     files.download(hitlog_file)
```
